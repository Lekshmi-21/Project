```python
df['trestbps']=df['trestbps'].fillna(df['trestbps'].mode()[0])
df['restecg']=df['restecg'].fillna(df['restecg'].mode()[0])
df['thalach']=df['thalach'].fillna(df['thalach'].mode()[0])
df['ca']=df['ca'].fillna(df['ca'].mode()[0])
df['thal']=df['thal'].fillna(df['thal'].mode()[0])
df.isna().sum()
```

```python
x=df.iloc[:,:-1].values
x
```

```python
y=df.iloc[:,-1].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=42)
```

```python
from sklearn.preprocessing import StandardScaler
Scaler=StandardScaler()
Scaler.fit(x_train)
x_train=Scaler.transform(x_train)
x_test=Scaler.transform(x_test)
```

```python
from sklearn.svm import SVC
model=SVC()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
y_test
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,ConfusionMatrixDisplay
acc=accuracy_score(y_test,y_pred)
acc
```

```python
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
cmd=ConfusionMatrixDisplay.from_predictions(y_test,y_pred)
cmd
```
